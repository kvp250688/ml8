## Problem Statement 

   In this assignment students have to find the frequency of words in a webpage. User can 
   use urllib and BeautifulSoup to extract text from webpage. 

```python
# Importing the nltk package
import nltk
%matplotlib inline
```

```python
# connecting to the website and checking the HTML data
import urllib.request
response = urllib.request.urlopen('http://php.net/')
htmlcontent = response.read()
print (htmlcontent)
```

```python
# Reading the data from webpage using the BeautifulSoup keyword, which will clean the HTML tags
from bs4 import BeautifulSoup
soupinfo = BeautifulSoup(htmlcontent,"html5lib")
text = soupinfo.get_text(strip=True)
print (text)
```

```python
# Spliting the words based on space between them
words = [t for t in text.split()] 
print (words)
```

## Word count with stop words

```python
# Calculating the frequency of words and printing it
freq = nltk.FreqDist(words) 
for word,count in freq.items(): 
    print (str(word) + ':' + str(count))
```

```python
# PLotting the first 40 words which are occurred most number of times
freq.plot(40, cumulative=False)
```

```python
print("the word which is repeated Maximum number of times:",freq.max())
```

## Removing Stop words and checking the count

```python
# Importing the stopwords module
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
listofWords = words[:] 
for word in listofWords:
    if word in stopwords.words('english'):
        listofWords.remove(word)
```

```python
# After removing the stop words and Calculating the frequency of words and printing it
freq_excludeSW = nltk.FreqDist(listofWords) 
for word_exSW,count_exSW in freq_excludeSW.items(): 
    print (str(word_exSW) + ':' + str(count_exSW))
```

```python
# PLotting the first 40 words which are occurred most number of times
freq.plot(40, cumulative=False)
```

```python
print("After removing the stop words, the word which is repeated Maximum number of times:",freq_excludeSW.max())
```

## Applying the Stemming and checking word frequency

```python
from nltk.stem.porter import PorterStemmer
import string
stemmer  = PorterStemmer()
punctuation = string.punctuation
stopwords   = set(nltk.corpus.stopwords.words('english'))


def normalize(text):
    for word in text:
        word = word.lower()
        if word in stopwords and word in punctuation:
            continue
        word = stemmer.stem(word)
        yield word

stemwords=list(normalize(words))
print(stemwords)
```

```python
# After Stemming, Calculating the frequency of words and printing it
freq_stem = nltk.FreqDist(stemwords) 
for word_stem,count_stem in freq_stem.items(): 
    print (str(word_stem) + ':' + str(count_stem))
```

```python
# PLotting the first 40 words which are occurred most number of times
freq_stem.plot(40, cumulative=False)
```

## Applying the lemmatizer and checking word frequency

```python
from nltk.stem.wordnet import WordNetLemmatizer
import string
from nltk.corpus import wordnet as wn
lemmatizer  = WordNetLemmatizer()
punctuation = string.punctuation
stopwords   = set(nltk.corpus.stopwords.words('english'))


def normalize(text):
    for word in text:
        word = word.lower()
        if word in stopwords and word in punctuation:
            continue
        word = lemmatizer.lemmatize(word)
        yield word

lemwords=list(normalize(words))
print(lemwords)
```

```python
# After Lemmatization, Calculating the frequency of words and printing it
freq_lem = nltk.FreqDist(lemwords) 
for word_lem,count_lem in freq_lem.items(): 
    print (str(word_lem) + ':' + str(count_lem))
```

```python
# PLotting the first 40 words which are occurred most number of times
freq_lem.plot(40, cumulative=False)
```
